```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
sector_count = df['Secteur'].value_counts()
sns.set(style="darkgrid")
sns.barplot(sector_count.index, sector_count.values, alpha=0.9)
plt.title('Frequency Distribution of Setors')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Sector', fontsize=12)
plt.show()
```

```python
labels = df['Secteur'].astype('category').cat.categories.tolist()
counts = df['Secteur'].value_counts()
sizes = [counts[var_cat] for var_cat in labels]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True) #autopct is show the % on plot
ax1.axis('equal')
plt.show()
```
